# Imputing Missing Data
* using forward fill for temps
* Interpolation for precipitation
* snow_depth with 0 as some regions may have no snowfall at all, so its hard to guess what is accurate here.

```python
df['max_temp'].fillna(method='ffill',inplace=True)
df['mean_temp'].fillna(method='ffill',inplace=True)
df['min_temp'].fillna(method='ffill',inplace=True)
```

```python
df['precipitation'].fillna(0,inplace=True)
```

```python
df['snow_depth'].fillna(0,inplace=True)
```

```python
df.isna().sum()
```

# Q1: What month had the most precipitation between 2000 through 2010?

```python
precip_monthly=df['precipitation'].resample('M').sum()
```

```python
max_date=df['precipitation'].idxmax()
```

```python
max_precip=df['precipitation'].max()
```

```python
max_month=precip_monthly['2000':'2010'].idxmax()
```

```python
max_value=precip_monthly['2000':'2010'].max()
```

```python
fig,ax=plt.subplots(figsize=(12,3))
ax.plot(precip_monthly.index,precip_monthly.values)
ax.axvline(max_month,color='r',linestyle='--',
            label=f"Max Precipitation\n({max_month.strftime('%B %Y')},{max_value:.2f} in)")
ax.axvline(max_date, color='g', linestyle='--',
            label=f"Date of Max Precipitation\n({max_date.strftime('%B %d, %Y')}, {max_precip:.2f} in)")
ax.legend()
ax.set_xlabel('Date')
ax.set_ylabel('Inches')
ax.set_title('Precipitation for 2000-2010')

# Major x-axis ticks
years_locator=mdates.YearLocator(base=1)
years_formatter=mdates.DateFormatter('%Y')
ax.xaxis.set_major_locator(years_locator)
ax.xaxis.set_major_formatter(years_formatter)

#Minor x-axis ticks
months_locator=mdates.MonthLocator(interval=3)
ax.xaxis.set_minor_locator(months_locator)
plt.minorticks_on()

plt.xticks(rotation=90)
ax.grid(True)
plt.tight_layout()
plt.show()
```

# Q2: Which year between 2000-2020 had the coolest average temperature?

```python
avg_temp_year=df['mean_temp'].resample('Y').mean()
min_temp_year=avg_temp_year.idxmin().year
min_temp_value=avg_temp_year.min()
```

```python
fig,ax=plt.subplots(figsize=(12,3))
ax.plot(avg_temp_year.index,avg_temp_year.values)
ax.axvline(pd.to_datetime(str(min_temp_year)),color='r',linestyle='--',
           label=f"Lowest Mean Temp\n({min_temp_year}, {min_temp_value:.2f} Degrees)")

ax.set_xlabel('Date')
ax.set_ylabel('Degrees')
ax.set_title('Average Temperature')

# Major x-axis ticks
years_locator=mdates.YearLocator(base=5)
years_formatter=mdates.DateFormatter('%Y')
ax.xaxis.set_major_locator(years_locator)
ax.xaxis.set_major_formatter(years_formatter)

# Minor x-axis ticks
years_minor_locator=mdates.YearLocator()
ax.xaxis.set_minor_locator(years_minor_locator)

ax.legend()
ax.grid(True)
plt.tight_layout()
fig.autofmt_xdate()
plt.xticks(rotation=90)
plt.show()
```
